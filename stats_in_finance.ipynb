```python
import pandas as pd

df = pd.DataFrame({
    'A': [4, 6, 3, 9, 15, 2],
    'B': [80, 105, 19, 90, 167, 182],
    'C': [0.3, 0.5, 0.43, 0.49, 0.36, 0.27],
    'D': [6000, 5900, 5700, 5300, 5500, 4800]
})

df.plot()
```

```python
# Calculate the percentage return of each column
returns = df.pct_change() * 100

# Drop the first row as it is na
returns = returns.dropna()

# Print the percentage returns
returns.plot()
```

```python
# Main computation functions
import math

# You could normally do df.std(), however the values are different than the supplied excel
def calculate_volatility(data, column):
    df = data[column]
    mean = df.mean()
    df['vs_mean'] = df - df.mean()
    df['deviation'] = df['vs_mean'] ** 2
    variance = df['deviation'].sum() / (len(df['deviation']))
    return math.sqrt(variance)

def calculate_corrolation(df):
    return df.corr()

def calculate_r_squared(df):
    return df.corr() ** 2
```

```python
calculate_volatility(returns, 'A')
```

```python
calculate_corrolation(returns)
```

```python
calculate_r_squared(returns)
```
